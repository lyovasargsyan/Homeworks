```python
import numpy as np
```

```python
def mean(x):
    return x.sum() / len(x)

def var(x):
    return ((x - mean(x)) ** 2).sum() / len(x)
    # return mean(x ** 2) - mean(x) ** 2

def std(x):
    return var(x) ** 0.5

def cov(x, y):
    return ((x - mean(x)) * (y - mean(y))).sum() / len(x)
    # return mean(x * y) - mean(x) * mean(y)

def corr(x, y):
    return cov(x, y) / (std(x) * std(y))
```

```python
X = np.array([5, 8, 12, 7, 8])
Y = np.array([23, 21, 25, 26, 25])
```

```python
mean(X), mean(Y)
```

```python
var(X), var(Y)
```

```python
cov(X, Y)
```

```python
corr(X, Y)
```

```python
X.mean(), Y.mean()
```

```python
X.var(), Y.var()
```
