### Logistic regression from scratch

```python
class MyLogisticRegression:
    
    def __init__(self, learning_rate=0.01, num_iterations=1000, verbose=False):
        self.learning_rate = learning_rate
        self.num_iterations = num_iterations
        self.verbose = verbose
        self.w = None
        self._costs = []
    
    def _sigmoid(self, z):
        # Ограничиваем z, чтобы избежать overflow в exp
        z = np.clip(z, -500, 500)
        return 1 / (1 + np.exp(-z))
    
    def _loss(self, yp, y):
        # Добавляем epsilon, чтобы не было log(0)
        epsilon = 1e-15
        yp = np.clip(yp, epsilon, 1 - epsilon)
        return (-y * np.log(yp) - (1 - y) * np.log(1 - yp)).mean()
    
    def fit(self, x, y):
        # Приводим y к одномерному массиву для стабильности
        y = np.asarray(y).flatten()
        
        intercept = np.ones((x.shape[0], 1))
        x_full = np.concatenate((intercept, x), axis=1)
        
        self.w = np.zeros(x_full.shape[1])
        self._costs = []
        
        for i in range(self.num_iterations):
            # Считаем предсказание
            z = x_full @ self.w
            yp = self._sigmoid(z)
            
            # Считаем лосс
            loss = self._loss(yp, y)
            self._costs.append(loss)
            
            # Считаем градиент
            gradient = (x_full.T @ (yp - y)) / y.shape[0]
            
            # Обновляем веса
            self.w -= self.learning_rate * gradient
            
            if self.verbose and i % 100 == 0:
                print(f'Iteration {i}, loss: {loss:.4f}')
    
    def predict_prob(self, x):
        intercept = np.ones((x.shape[0], 1))
        x_full = np.concatenate((intercept, x), axis=1)
        return self._sigmoid(x_full @ self.w)
    
    def predict(self, x):
        # Стандартный порог 0.5
        return (self.predict_prob(x) >= 0.5).astype(int)

    def get_costs(self):
        return np.array(self._costs)
```

```python
log_reg = MyLogisticRegression()
```

```python
log_reg.fit(x_train, y_train)
```

```python
p_probs = log_reg.predict_prob(x_test)
p_probs.round(3)
```

```python
p = log_reg.predict(x_test)
p
```

```python
np.sum(y_test == p) / y_test.shape[0]
```

```python
log_reg.get_costs()
```

```python
fig = plt.figure(figsize=(10, 6))

plt.plot(np.arange(1000), log_reg.get_costs())

plt.title("Development of cost over training")
plt.xlabel("Number of iterations")
plt.ylabel("Cost")
```

```python
y_p_train = log_reg.predict(x_train)
y_p_test = log_reg.predict(x_test)

print(f"train accuracy: {100 - np.mean(np.abs(y_p_train - y_train)) * 100}%")
print(f"test accuracy: {100 - np.mean(np.abs(y_p_test - y_test))}%")
```

### K- Fold Cross Validation

```python
from sklearn.model_selection import KFold
```

```python
scores = []
cv = KFold(n_splits=10, random_state=42, shuffle=True)
for train_index, test_index in cv.split(x):
    best_svr = LogisticRegression(random_state=0)
    print("Train Index: ", train_index, "\n")
    print("Test Index: ", test_index)
    x_train, x_test, y_train, y_test = x[train_index], x[test_index], y[train_index], y[test_index]
    best_svr.fit(x_train, y_train)
    scores.append(best_svr.score(x_test, y_test))
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scores_train = []
scores_test = []
cv = KFold(n_splits=5, shuffle=True, random_state=42)
for train_ind, test_ind in cv.split(x):
    x_train, x_test, y_train, y_test = x[train_ind], x[test_ind], y[train_ind], y[test_ind]
    sc = StandardScaler()
    x_train = sc.fit_transform(x_train)
    x_test = sc.transform(x_test)
    model = LogisticRegression(C=0.05,solver="newton-cg")
    model.fit(x_train, y_train)
    acc_test = accuracy_score(y_test, model.predict(x_test))
    acc_train = accuracy_score(y_train, model.predict(x_train))
    scores_test.append(acc_test)
    scores_train.append(acc_train)
print(scores_train)
print(scores_test)
np.mean(scores_train), np.mean(scores_test)
```
